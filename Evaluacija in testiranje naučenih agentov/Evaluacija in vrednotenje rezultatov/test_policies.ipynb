## Train the agent and evaluate it

```python
for ii, params in enumerate(params_grid):
  
  algorithm_logger = AlgorithmLogger(
    max_no_of_episodes=200,
    max_time_steps_per_episode=500,
    observation_size=7 if env_name == 'CartPole-v3' else 4,
    goal=1.0)
  
  experiment_name = model_name + \
                    '_plr' + str(params['pi_lr']) + \
                    '_vlr' + str(params['vf_lr']) + \
                    '_spe' + str(params['steps_per_epoch']) + \
                    '_gam' + str(params['gamma']) + \
                    '_lam' + str(params['lam']) + \
                    '_tkl' + str(params['target_kl']) + \
                    '_tpi' + str(params['train_pi_iters']) + \
                    '_tvi' + str(params['train_v_iters']) + \
                    '_cr' + str(params['clip_ratio']) + \
                    '_seed' + str(params['seed'])

  model_logdir = os.path.join('/home/anej/Downloads/alt/content/spinningup/data/alt', experiment_name)
  logger_kwargs = dict(output_dir=model_logdir, exp_name=experiment_name)

  print(f'On evaluation {ii + 1} / {len(params_grid)}.')

  env, get_action = load_policy_and_env(model_logdir, env_name,
                                        False, 'last', True)
  run_policy(env, get_action, 0, 200, False, algorithm_logger=algorithm_logger)

  for what in PlottingConstants.AVERAGE_EPISODE_STATISTICS:
    algorithm_logger.plot_summary(what)
```
